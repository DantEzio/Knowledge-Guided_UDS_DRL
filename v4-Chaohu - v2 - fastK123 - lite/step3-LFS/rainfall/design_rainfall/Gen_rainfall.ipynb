# Real rainfall

# Expand rainfall

```python
rr = np.load('test_raindata.npy').tolist()
```

```python
rains = {}
for id in range(len(rr)):
    tem = []
    for i in range(len(rr[id])):
        tem.append(float(rr[id][i][1])+100)
    rains[id] = tem
```

```python
def real_rain_op(data):    
    Hs = []
    delta,dura = 1, len(data)
    for i in range(dura//delta):
        Hs.append(data[i])
    tsd = Hs
    ts = []
    for i in range(dura//delta):
        if np.mod(i,5)==0:
            t = i*delta
            key = '01/01/2001 '+str(9+t//60).zfill(2)+':'+str(t % 60).zfill(2)+':'+'00'#'08/28/2015 '+str(9+t//60).zfill(2)+':'+str(t % 60).zfill(2)+':'+'00'
            ts.append([key,tsd[i]])
    return ts
```

```python
raindata=[]
for i in range(len(rains)):
    raindata.append(real_rain_op(rains[i]))
np.save('expanded_raindata.npy',np.array(raindata))
```
