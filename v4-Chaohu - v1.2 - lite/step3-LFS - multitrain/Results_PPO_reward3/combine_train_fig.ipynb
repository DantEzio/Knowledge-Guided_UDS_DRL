```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r1 = np.load('./Results/41_Train.npy',allow_pickle=True).tolist()
r2 = np.load('./Results/42_Train.npy',allow_pickle=True).tolist()
r3 = np.load('./Results/43_Train.npy',allow_pickle=True).tolist()
r4 = np.load('./Results/44_Train.npy',allow_pickle=True).tolist()
r5 = np.load('./Results/45_Train.npy',allow_pickle=True).tolist()
r6 = np.load('./Results/46_Train.npy',allow_pickle=True).tolist()
rs = np.load('./Results/1_Train.npy',allow_pickle=True).tolist()
```

```python
r = list(np.array(r1['Episode_reward'])/0.9)+list(np.array(r2['Episode_reward'])/0.9) \
        +r3['Episode_reward']+list(np.array(r4['Episode_reward'])*0.9) \
        +list(np.array(r5['Episode_reward'][:180])*0.9) \
        +list(np.array(r6['Episode_reward'])*0.9)
```

```python
# timely average
t = 3
for i in range(t,len(r)-t):
    r[i] = np.mean(r[i-t:i+t])
```

```python
plt.plot(r)
plt.plot(rs['Episode_reward'])
```

```python
rs['Episode_reward'] = r
np.save('./Results/4_Train.npy',rs)
```
